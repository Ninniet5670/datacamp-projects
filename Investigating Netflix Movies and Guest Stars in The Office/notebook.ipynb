```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/office_episodes.csv')

has_guests_df = df[df['has_guests'] == True]
not_guests_df = df[df['has_guests'] == False]

df.head()
```

```python
top_stars = has_guests_df[has_guests_df['viewership_mil'] == max(has_guests_df['viewership_mil'])]['guest_stars']
top_star = top_stars[77].split(',')[0]
top_star
```

```python
def return_colors(dataframe):
    colors = []
    for rating in dataframe['scaled_ratings']:
        if rating < 0.25:
            colors.append('red')
        elif rating < 0.50:
            colors.append('orange')
        elif rating < 0.75:
            colors.append('lightgreen')
        else:
            colors.append('darkgreen')
            
    return colors
```

```python
fig = plt.figure(figsize=(11, 7))
ax = fig.add_subplot(111)

ax.scatter(has_guests_df['episode_number'], has_guests_df['viewership_mil'], 
           c=return_colors(has_guests_df), s=250, marker='*', label='Episodes With Guest Stars')
ax.scatter(not_guests_df['episode_number'], not_guests_df['viewership_mil'], 
           c=return_colors(not_guests_df), s=25, marker='o', label='Episodes Without Guest Stars')

ax.set_title('Popularity, Quality, and Guest Appearances on the Office')
ax.set_xlabel("Episode Number")
ax.set_ylabel("Viewership (Millions)")
ax.legend()

plt.show()
```
